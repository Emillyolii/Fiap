Vladimir Sidorov desistiu de seu trabalho convencional na KGB e decidiu empreender em 2010 no ramo alimentício. Com suas economias abriu um foodtruck no bairro do Ipiranga – SP especializado na venda de Pirozhki, onde seu grande diferencial era trabalhar com ingredientes importados dando uma qualidade extraordinária ao produto. Após o crescimento das vendas, a modalidade delivery foi implementada para auxiliar a empresa. Vladimir era extremamente popular no bairro, todos o conheciam, mas em 2019 se mudou para um ponto melhor e não tivemos mais notícia. Os últimos dados compartilhados das vendas de Vlad foram disponibilizados antes de sua partida.


##1.   Dado o problema acima, qual o melhor algoritmo para descobrir os dados de 2015?

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Dados de vendas
dados = {
    'Ano': [2010, 2011, 2012, 2014, 2017, 2018, 2019],
    'Vendas': [723, 814, 905, 1087, 1360, 1451, 1542]
}

# Crie um DataFrame a partir dos dados
df = pd.DataFrame(dados)

# Imprima o DataFrame para verificação
print(df)
```

```python
# Divida os dados em conjuntos de treinamento (até 2018) e teste (2019)
X_treino = df[df['Ano'] <= 2018]['Ano'].values.reshape(-1, 1)
y_treino = df[df['Ano'] <= 2018]['Vendas'].values
X_teste = df[df['Ano'] == 2019]['Ano'].values.reshape(-1, 1)
y_teste = df[df['Ano'] == 2019]['Vendas'].values
# Crie um objeto de regressão linear
modelo_regressao = LinearRegression()

# Treine o modelo com os dados de treinamento
modelo_regressao.fit(X_treino, y_treino)

# Faça previsões para as vendas de 2015
vendas_2015_previstas = modelo_regressao.predict(np.array([[2015]]))

# Imprima o valor previsto para 2015
print(f"Previsão de vendas para 2015: {vendas_2015_previstas[0]:.2f}")
```

```python
# Plotar os dados de treinamento e a linha de regressão
plt.scatter(X_treino, y_treino, color='blue', label='Dados de Treinamento')
plt.plot(X_treino, modelo_regressao.predict(X_treino), color='red', label='Regressão Linear')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.title('Regressão Linear - Dados de Treinamento')
plt.legend()
plt.show()
```

```python
# Imprima o coeficiente angular (inclinação da linha de regressão)
coeficiente_angular = modelo_regressao.coef_[0]
print(f"Coeficiente Angular (a): {coeficiente_angular:.2f}")
```

```python
# Faça a previsão para 2020
vendas_2020_previstas = modelo_regressao.predict(np.array([[2020]]))

# Imprima o valor previsto para 2020
print(f"Previsão de vendas para 2020: {vendas_2020_previstas[0]:.2f}")
```

```python
# Faça a previsão para 2021
vendas_2021_previstas = modelo_regressao.predict(np.array([[2021]]))
print(f"Previsão de vendas para 2021: {vendas_2021_previstas[0]:.2f}")
```
